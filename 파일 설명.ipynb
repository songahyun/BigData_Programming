## 이 파일은 각 파일을 설명하는 파일입니다.

### 파일 설명

**지표 생성(인구, 교통, 문화, 의료)**  
인구밀도, 교통 편의성, 문화 인프라 지수, 의료 인프라 지수를 생성하는 코드 파일  
(문화 인프라 지수는 관악구의 누락으로 인해 사용하지 못하기 때문에 문화공간 데이터를 새로 수집하여 예술인프라지수를 생성함)

**지표 생성(예술)**  
자치구별 미술관, 박물관의 수를 합산한 뒤 인구 수로 나누는 코드 파일

**지표 생성(상권변화지표)**  
상권변화지표에 대한 설명이 포함되어 있는 파일  
상권변화지표를 숫자형 데이터로 변경하고 성장가능성이라는 변수로 만든 파일
  
**변수 통합**  
인구밀도, 교통 편의성, 문화 인프라 지수, 예술 인프라 지수, 의료 인프라 지수, 상권변화지표를 통합하는 코드 파일  
이후 통합한 데이터셋을 통한 간단한 분석을 수행했지만 이 부분은 발표와 피피티에 포함하지 않았으므로 꼭 확인하실 필요는 없습니다.  
  
**회귀분석**
변수 통합_original.csv를 사용하여 회귀분석과 로지스틱 회귀, DBSCAN을 진행한 코드  
단, 발표와 피피티에는 회귀분석 결과만 사용했습니다.
  
**지표 생성(학교)**  
자치구별 초/중/고등학교 수를 계산하고 합산하는 코드 파일

**학교 변수 통합**  
기존 모든 지표를 통합한 파일(변수 통합_original.csv)에 학교 변수도 추가하는 코드 파일  
이후 간단한 분석을 수행했지만 이 부분 역시 발표와 피피티에 포함하지 않았으므로 꼭 확인하실 필요는 없습니다.  

**회귀분석2**  
학교 변수가 추가된 데이터셋(변수 통합_original2.csv)을 사용하여 회귀분석과 로지스틱 회귀을 진행한 코드  
단, 발표와 피피티에는 회귀분석 결과만 사용했습니다.
