### β— λ¬Έν™” μΈν”„λΌ μ§€μλ” NaN κ°’ ν¬ν•¨ β†’ μμ μΈν”„λΌμ§€μλ΅ λ€μ²΄

```python
# μ‹κ°ν™” μ¤€λΉ„
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# κ²½κ³  λ¬΄μ‹
warnings.filterwarnings("ignore")

# ν•κΈ€ ν°νΈ μ„¤μ • (μλ„μ° κΈ°μ¤€)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import shap

# 1. λ°μ΄ν„° μ¤€λΉ„
df = df_merge.copy()

# 2. μ‚¬μ©ν•μ§€ μ•μ„ μ»¬λΌ μ κ±° λ° ν•„μ”ν• μ»¬λΌλ§ μ„ νƒ
# - λ¬Έν™” μΈν”„λΌ μ§€μλ” κ΄€μ•…κµ¬ NaN μμΌλ―€λ΅ μ μ™Έ
# - μƒκ¶_λ³€ν™”_μ§€ν‘λ” λ¬Έμμ—΄ν•μ΄λ―€λ΅ ν„μ¬ λ¶„μ„μ—μ„λ” μ‚¬μ©ν•μ§€ μ•μ
X = df[['μΈκµ¬λ°€λ„', 'κµν†µ νΈμμ„±', 'μμ μΈν”„λΌμ§€μ', 'μλ£ μΈν”„λΌ μ§€μ']]
y = df['μ„±μ¥κ°€λ¥μ„±']

# 3. NaN μ²λ¦¬
# - λ§μ•½ μμ μΈν”„λΌμ§€μμ—λ„ NaNμ΄ μμ„ κ²½μ° μ κ±°
X = X.dropna()
y = y.loc[X.index]  # Xμ™€ yμ μΈλ±μ¤ μΌμΉμ‹ν‚΄

# 4. ν‘μ¤€ν™”
# - κ° λ³€μ λ‹¨μ„κ°€ λ‹¤λ¥΄λ―€λ΅ ν‰κ·  0, ν‘μ¤€νΈμ°¨ 1λ΅ ν‘μ¤€ν™”
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. ν•™μµ/κ²€μ¦ λ¶„λ¦¬
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# ===============================
# β¬› 1. λ‹¤ν•­ λ΅μ§€μ¤ν‹± νκ·€ λ¶„μ„
# ===============================
logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=42)
logreg.fit(X_train, y_train)

# κ³„μ ν™•μΈ (μ„¤λ…λ ¥ λ¶„μ„)
coeff_df = pd.DataFrame(logreg.coef_, columns=X.columns)
coeff_df['ν΄λμ¤'] = logreg.classes_
print("π“ [λ‹¤ν•­ λ΅μ§€μ¤ν‹± νκ·€ κ³„μ]")
print(coeff_df.set_index('ν΄λμ¤'))

# μ •ν™•λ„ λ³΄κ³ 
print("\nπ“ [λ¶„λ¥ μ •ν™•λ„]")
y_pred = logreg.predict(X_test)
print(classification_report(y_test, y_pred))

# =====================================
# β¬› 2. κ²°μ •νΈλ¦¬ κΈ°λ° Feature Importance
# =====================================
tree = DecisionTreeClassifier(max_depth=4, random_state=42)
tree.fit(X_train, y_train)

# μ‹κ°ν™”
feat_imp = pd.Series(tree.feature_importances_, index=X.columns)
feat_imp.sort_values().plot(kind='barh', title='π― Feature Importance (Decision Tree)', figsize=(7,4))
plt.tight_layout()
plt.show()

# ========================
# β¬› 3. SHAP κΈ°λ° μ„¤λ… λ¶„μ„
# ========================
explainer = shap.Explainer(tree, X_train)
shap_values = explainer(X_test)

# summary plot
shap.summary_plot(shap_values, features=X_test, feature_names=X.columns)
```

```python
# SHAP ννΈλ§µ
import shap
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ν΄λμ¤ 0 (μ„±μ¥κ°€λ¥μ„± 1)μ— λ€ν• interaction κ°’ μ¶”μ¶
interaction_values = shap_values[0]  # ν΄λμ¤ λ²νΈ 0: μ„±μ¥κ°€λ¥μ„±=1

# λ³€μ μ΄λ¦„ λ¦¬μ¤νΈ
feature_names = X.columns

# ν‰κ·  interaction value ν–‰λ ¬ μƒμ„± (μ λ“κ°’ κΈ°μ¤€μΌλ΅ μν–¥λ ¥ λ³΄κΈ° μ„ν•΄ ν‰κ·  μ λ“κ°’ μ‚¬μ©)
interaction_mean = np.abs(interaction_values).mean(axis=0)

# DataFrameμΌλ΅ λ³€ν™
interaction_df = pd.DataFrame(interaction_mean, columns=feature_names, index=feature_names)

# ννΈλ§µ μ‹κ°ν™”
plt.figure(figsize=(10, 8))
sns.heatmap(interaction_df, cmap='viridis', annot=True, fmt=".3f", square=True, cbar_kws={"label": "Mean |SHAP interaction|"})
plt.title("SHAP λ³€μ κ°„ μƒνΈμ‘μ© ννΈλ§µ (μ„±μ¥κ°€λ¥μ„±=1)")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
